```python
from scipy import special
import numpy as np

import matplotlib.pyplot as plt
from matplotlib import cm, colors
from mpl_toolkits.mplot3d import Axes3D
```

```python
num = 500
N = 3
S = 2
X = np.linspace(-S, S, num)
Y = np.linspace(-S, S, num)
X, Y = np.meshgrid(X, Y)

R = np.sqrt(X**2 + Y**2)
D = np.arctan2(Y, X)

P = np.pow(R, N) * np.cos(N * D)

K = - N**2 * (N - 1)**2 * np.pow(R, 2 * N) / (R**2 + N**2 * np.pow(R, 2 * N))**2
```

```python
fig = plt.figure(dpi=300)
ax = fig.add_subplot(111, projection='3d')
ax.set_box_aspect((2, 2, 3))
ax.set_aspect(aspect='equal')

cmap = cm.jet
norm = colors.Normalize(vmin = np.min(K), vmax = np.max(K))

ax.plot_surface(X, Y, P, 
                facecolors=cmap(norm(K)),
                # edgecolor='k',
                alpha=0.6,
                linewidth=0.3
                )

cbar = plt.colorbar(cm.ScalarMappable(norm=norm,cmap=cmap),ax=ax,pad=0.1)
# Labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
# ax.set_title('3D Parametric Surface: Sphere')
# Show the plot
plt.show()
```
